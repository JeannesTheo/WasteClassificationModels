```python
import random

import numpy as np
import pandas as pd
import shutil
from roboflow import Roboflow
import os
import re
```

```python
def clean_up(integral=False):
    if os.path.exists('GARBAGE-CLASSIFICATION-3-2'):
        shutil.rmtree('GARBAGE-CLASSIFICATION-3-2')
    if integral:
        if os.path.exists('data'):
            shutil.rmtree('data')
    else:
        for file in os.listdir('data'):
            file = os.path.join('data', file)
            if os.path.isfile(file):
                os.remove(file)

def move_files_recursively(src,dest='data',files=None):
    if not os.path.exists(dest):
        os.makedirs(dest)
    files_to_move = []
    if files is None:
        for root, dirs, files in os.walk(src):
            for file in files:
                if file.endswith('.jpg'):
                    files_to_move.append(os.path.join(root, file))
    else:
        interesting_files = list(set(files) & set(os.listdir(src)))
        for file in interesting_files:
            files_to_move.append(os.path.join(src, file))
    for file in files_to_move:
        shutil.move(file, dest)
```

```python
data_quantity = 920
test_quantity = 250

def remove_files_randomly(directory_path, target_files):
    """
    Remove files randomly from a directory until the desired count is achieved.

    Parameters:
    - directory_path: The path to the directory.
    - target_files: The desired number of files to keep.
    """
    files = os.listdir(directory_path)

    if len(files) > target_files:
        excess_files = len(files) - target_files
        for i in range(excess_files):
            file_to_remove = os.path.join(directory_path, files[random.randint(0,len(files)-1)])
            os.remove(file_to_remove)
            files.remove(os.path.basename(file_to_remove))
for i in classes:
    remove_files_randomly(f'content/all_data/data/{i}',data_quantity)
    remove_files_randomly(f'content/all_data/test/{i}',test_quantity)
```
